# Lecture 7 demonstration on basic plotting

Using the lecture, we provided a brief demo of how a function is plotted on the Jupyter notebook. Remember that cells or commands on the notebook are executed by typing `Shift-Enter` or `Ctrl-Enter`

```python
# Load in basic array/matrix packages and plotting packages
import numpy as np # This imports numpy, a package for manipulating vectors and matrices
import matplotlib.pyplot as plt # This imports matplotlib's plotter
```

```python
# Create a vector of 20 elements from 0 to 2*pi with equally spaced points
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('x'); 
```
