# Lecture 10: Euler's method to solve IVPs

We apply our own implementation of Euler's method to solve the previous IVP.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp # solver for IVPs
```

```python
tmax = 2                # max time
dt = 0.2                # time step
N = int(tmax/dt)        # number of steps (converted to integer)

# Our ODE function
def myfunc(t, Y):
    y, yp = Y
    ypp = -1/(1 + ep*y)**2
    return np.array([yp, ypp])
    
# Set the value of epsilon within the function
ep = 0.2

# Initial condition
Y0 = [0, 1]

# Start with initial condition
Y = Y0
t = 0
plt.plot(t, Y0[0], 'ro')
for i in range(0, N):
    # Euler step
    Y = Y + myfunc(t, Y)*dt
    # Increment time
    t = t + dt
    plt.plot(t, Y[0], 'bo')

# Let's compare with the built-in solver
mysol = solve_ivp(myfunc, [0, tmax], Y0, dense_output=True)

# Create a dense mesh for plotting
tt = np.linspace(0, tmax, 100)
yy = mysol.sol(tt)
y0 = -1/2*tt**2 + tt
plt.plot(tt, yy[0], 'r')
plt.plot(tt, y0, '--'); 
plt.xlabel('time'); plt.ylabel('y');
plt.legend('Euler', 'solve_ivp', 'asymptotic')
```
