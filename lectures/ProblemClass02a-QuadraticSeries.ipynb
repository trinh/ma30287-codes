The intention of this notebook is to test the asymptotic expansion of the roots of the quadratic
$$
\epsilon x^2 + x - 1 = 0,
$$
studied in the lectures. You will compare the exact solution values, computed using the quadratic equation, with the asymptotic expansion.

```python
# Load in packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an array of epsilon values
ep = np.linspace(1, 0.1, 50)
root1 = (-1 + np.sqrt(1 + 4*ep))/(2*ep)
root2 = (-1 - np.sqrt(1 + 4*ep))/(2*ep)
asym1 = 1 - ep;
asym2 = -1/ep - 1;
plt.plot(ep, root1, 'o')
plt.plot(ep, root2, 'o')
plt.plot(ep, asym1, '-')
plt.plot(ep, asym2, '-')
plt.xlabel('epsilon'); plt.ylabel('root');
```

```python
# An alternative visualisation is to examine the error: error = exact - approximation
plt.plot(ep, root1 - asym1, 'o-')
plt.xlabel('eps'); plt.ylabel('exact - approximation');
plt.title('Convergence of the error for the first root');
plt.grid()
```

```python
# An alternative visualisation is to examine the error: error = exact - approximation
plt.plot(ep, root2 - asym2, 'o-')
plt.xlabel('eps'); plt.ylabel('exact - approximation');
plt.title('Convergence of the error for the second root');
plt.xlim([0, 1]); plt.ylim([0, 0.4])
plt.grid()
```
