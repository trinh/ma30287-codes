# Chapter 8: plotting the solution of an IVP

This script solves the IVP $\epsilon T' = R(t) - T$ with the intiial condition $T(0)= T_s$. It then compares the numerical solution with the asymptotic solution. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
Ts = 0
ep = 0.9
A = 0.1
f = lambda t, T: 1/ep*(1 + A*np.cos(t) - T)

tmax = 20
sol = solve_ivp(f, [0, tmax], [Ts], dense_output=True)

t = np.linspace(0, tmax, 1000)
y = sol.sol(t)
plt.plot(t, y[0], '-')
plt.plot(t, 1 + A*np.cos(t), '--')
plt.plot(t, 1 + A + (Ts - (1 + A))*np.exp(-t/ep), '--')
plt.grid(True)
```
