# PS6 solutions supplement

This code plots some figures relevant to Q4 of the problem set. During the lectures, we investigated the behaviour of the Stommel box model under changes in the lambda parameter (refer to Lecture 31 notes). The difference with this set is to examine differences in the steady-state solutions when the R parameter is changed instead.

We can adjust the phi definition in the below code to allow R to be used as a parameter. Once that is done, we simply plot for a few different values of R. 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import root

delt = 1/6
lam = 0.2
R = 2
    
def phi(f, R):
    phi = delt*R/(delt + np.abs(f)) - 1/(1 + np.abs(f))
    return phi

fmat = np.linspace(-2.5, 2.5, 101)
fig = plt.figure(figsize=(8, 8))
plt.plot(fmat, phi(fmat, 1), 'r', label='$\phi$, $R = 1$')
plt.plot(fmat, phi(fmat, 2), 'g', label='$\phi$, $R = 2$')
plt.plot(fmat, phi(fmat, 3), 'm', label='$\phi$, $R = 3$')
plt.plot(fmat, lam*fmat, 'b', label='$\lambda f$' )
plt.ylim((-1.5,2))
plt.xlabel("$f$"); plt.ylabel("$\phi$"); plt.legend(); plt.grid(1)
```
