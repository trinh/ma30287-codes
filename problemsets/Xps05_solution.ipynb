```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q2

```python
xmin = -3.5; xmax = 5.5;
x00 = np.linspace(xmin, xmax, 50);
S = lambda x: (x**3 - 3*x**2 - 9*x)/4
y00 = S(x00);

# Extrema
xe = [-1, 3];
ye = [5/4, -27/4];

# Intersections with manifold
xi = [5, -3];
yi = [5/4, -27/4];

plt.plot(x00, y00)
plt.plot(xe, ye, 'o')
plt.plot(xi, yi, 'o')
plt.grid(1)
plt.xlabel('x');
plt.ylabel('y');
```
